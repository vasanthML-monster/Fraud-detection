## Json parsing

```python
#importing required libraries
import pandas as pd
import numpy as np
import json
import matplotlib.pylab as plt
import scipy as sp
```

```python
# read file
JsonList=[]
for line in open('E:\summit minds\customersdata.json', 'r'):
    JsonList.append(json.loads(line))
```

```python
#parse file 
JsonList[0]
```

```python
def extract_values(obj, key):
    """Pull all values of specified key from nested JSON."""
    arr = []

    def extract(obj, arr, key):
        """Recursively search for values of key in JSON tree."""
        if isinstance(obj, dict):
            for k, v in obj.items():
                if isinstance(v, (dict, list)):
                    extract(v, arr, key)
                elif k == key:
                    arr.append(v)
        elif isinstance(obj, list):
            for item in obj:
                extract(item, arr, key)
        return arr

    results = extract(obj, arr, key)
    return results

```

## Customer and order details

```python

Data_cust_Orders=pd.DataFrame(columns=['customerEmail','customerPhone','customerDevice','customerIPAddress','customerBillingAddress','orderId','orderAmount','orderState','orderShippingAddress'])


Rcols=['orderId','orderAmount','orderState','orderShippingAddress']
Scols=['customerEmail','customerPhone','customerDevice','customerIPAddress','customerBillingAddress','fraudulent']
flag='orders'
for num in range(len(JsonList)):
    Orders=pd.DataFrame()
    for i in Rcols:
        val=extract_values(JsonList[num][flag],i)
        Orders[i]=val
    for k in Scols:
        val=extract_values(JsonList[num],k)
        Orders[k]=np.nan
        Orders[k]=Orders[k].fillna(str(val[0]))
    Data_cust_Orders=Data_cust_Orders.append(Orders,ignore_index=True)

Data_cust_Orders.head()
```

```python
Data_cust_Orders.to_csv('E:\summit minds\Data_cust_Orders.csv')
```
